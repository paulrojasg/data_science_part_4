```python
from connection_script import connect_databases
import pandas as pd
import locale
```

```python
db_op, db_etl = connect_databases()
```

## EXTRACT

```python
sales_order_detail = pd.read_sql_query("SELECT * FROM [Sales].[SalesOrderDetail]", db_op)
sales_order_detail
```

```python
sales_order_header = pd.read_sql_query("SELECT * FROM [Sales].[SalesOrderHeader]", db_op)
sales_order_header
```

```python
sales_customer = pd.read_sql_query("SELECT * FROM [Sales].[Customer]", db_op)
sales_customer
```

```python
sales_store = pd.read_sql_query("SELECT * FROM [Sales].[Store]", db_op)
sales_store
```

```python
product = pd.read_sql_query("SELECT ProductID,ProductNumber FROM [Production].[Product]", db_op)
product
```

```python
currency_rate = pd.read_sql_query("SELECT CurrencyRateID,ToCurrencyCode FROM [Sales].[CurrencyRate]", db_op)
currency_rate
```

```python
# Extract dimensions
dim_currency = pd.read_sql_query("SELECT CurrencyKey,CurrencyAlternateKey FROM DimCurrency", db_etl)
dim_date = pd.read_sql_query("SELECT DateKey,FullDateAlternateKey FROM DimDate", db_etl)
dim_employee = pd.read_sql_query("SELECT EmployeeKey FROM DimEmployee", db_etl)
dim_product = pd.read_sql_query("SELECT ProductKey,ProductAlternateKey FROM DimProduct", db_etl)
dim_promotion = pd.read_sql_query("SELECT PromotionKey FROM DimPromotion", db_etl)
dim_reseller = pd.read_sql_query("SELECT ResellerKey FROM DimReseller", db_etl)
dim_sales_territory = pd.read_sql_query("SELECT SalesTerritoryKey FROM DimSalesTerritory", db_etl)
```

## TRANSFORM

```python
#Filter out online orders
merged_df = sales_order_detail.merge(sales_order_header, on='SalesOrderID')
merged_df = merged_df[merged_df['OnlineOrderFlag'] == 0] 

# Merge with dimensions
merged_df = merged_df.merge(sales_customer, on='CustomerID')

# Convert 'FullDateAlternateKey' to datetime
dim_date['FullDateAlternateKey'] = pd.to_datetime(dim_date['FullDateAlternateKey'])

# Now perform the merges
merged_df = merged_df.merge(dim_date.rename(columns={'DateKey':'OrderDateKey'}), left_on="OrderDate", right_on="FullDateAlternateKey", suffixes=('_merged', '_OrderDate'))
merged_df = merged_df.merge(dim_date.rename(columns={'DateKey':'DueDateKey'}), left_on="DueDate", right_on="FullDateAlternateKey", suffixes=('_merged', '_DueDate'))
merged_df = merged_df.merge(dim_date.rename(columns={'DateKey':'ShipDateKey'}), left_on="ShipDate", right_on="FullDateAlternateKey", suffixes=('_merged', '_ShipDate'))

merged_df[['OrderDateKey', 'DueDateKey', 'ShipDateKey', 'StoreID', 'SalesPersonID']]
```

```python
# Rename the index in sales_store and then reset it
sales_store.index.names = ['ResellerID']
sales_store_reset = sales_store.reset_index()

merged_df.drop(columns=['ModifiedDate_x','rowguid_x'], inplace=True)
# Merge merged_df with sales_store_reset
merged_df = merged_df.merge(sales_store_reset, left_on='StoreID', right_on='BusinessEntityID', how='left')
merged_df = merged_df.merge(dim_reseller, left_on='ResellerID', right_on='ResellerKey', how='left')
merged_df = merged_df.merge(dim_employee, left_on='SalesPersonID_x', right_on='EmployeeKey', how='left')
print(merged_df.columns)
merged_df[['ResellerKey', 'EmployeeKey']]
```

```python
merged_df[['ProductID', 'CurrencyRateID', 'TerritoryID_x', 'SpecialOfferID']]
#sales_order_header[sales_order_header['CurrencyRateID'].notnull()]
```

```python
dim_product = dim_product.merge(product, left_on='ProductAlternateKey', right_on='ProductNumber', how='left')
print(dim_product.duplicated(subset=['ProductAlternateKey']).sum())
dim_product = dim_product.drop_duplicates(subset=['ProductID'])
dim_product
```

```python
merged_df['CurrencyAlternateKey'] = merged_df['CurrencyRateID'].map(currency_rate.set_index('CurrencyRateID')['ToCurrencyCode'])
merged_df['CurrencyAlternateKey'].notnull().sum()
```

```python

merged_df = merged_df.merge(dim_product, left_on='ProductID', right_on='ProductID', how='inner')
merged_df = merged_df.merge(dim_currency, left_on='CurrencyAlternateKey', right_on='CurrencyAlternateKey', how='left')
merged_df = merged_df.merge(dim_promotion, left_on='SpecialOfferID', right_on='PromotionKey', how='left')
merged_df = merged_df.merge(dim_sales_territory, left_on='TerritoryID_x', right_on='SalesTerritoryKey', how='left')
#merged_df[['ProductKey', 'CurrencyKey', 'PromotionKey', 'SalesTerritoryKey']]
print(merged_df.columns)
print(merged_df['CurrencyRateID'].notnull().sum())
print(merged_df['CurrencyKey'].notnull().sum())
merged_df[['ProductKey', 'CurrencyKey', 'PromotionKey', 'SalesTerritoryKey', 'LineTotal']]
```

```python
#Drop the duplicate orders created when merging with Dim_Product
#merged_df = merged_df.drop_duplicates(subset='SalesOrderDetailID')
#merged_df
```

```python
#Number Items in each Sales Order
merged_df['SalesOrderLineNumber'] = merged_df.groupby('SalesOrderID').cumcount() + 1
#Adding 'SO' to the 'SalesOrderID' column
merged_df['SalesOrderID'] = merged_df['SalesOrderID'].apply(lambda x: 'SO'+str(x))

merged_df[['SalesOrderID', 'SalesOrderLineNumber']]
```

```python
# Calculate the 'DiscountAmount' and 'SalesAmount'
merged_df['DiscountAmount'] = merged_df['LineTotal'] * merged_df['UnitPriceDiscount']
#merged_df['SalesAmount'] = merged_df['SubTotal'] * (1 - merged_df['UnitPriceDiscount'])

merged_df[['UnitPrice', 'UnitPriceDiscount', 'DiscountAmount', 'SubTotal', 'LineTotal']]
```

```python
#Cost of each product
def get_product_standard_cost(row):
    query = "SELECT dbo.ufnGetProductStandardCost(?, ?) AS ProductStandardCost" #ProductID, OrderDate => Product Cost on given date
    params = (row['ProductID'], row['OrderDate'])
    return pd.read_sql_query(query, db_op, params=params).iloc[0, 0]

merged_df['ProductStandardCost'] = merged_df.apply(get_product_standard_cost, axis=1)
merged_df['TotalProductCost'] = merged_df['ProductStandardCost'] * merged_df['OrderQty']

merged_df[['ProductStandardCost', 'TotalProductCost']]
```

```python
# Adding key for the null values in 'CurrencyKey' column
merged_df['CurrencyKey'] = merged_df['CurrencyKey'].fillna(100)
```

```python
result = merged_df[['ProductKey', 'OrderDateKey', 'DueDateKey', 'ShipDateKey', 'ResellerKey', 'EmployeeKey', 'PromotionKey', 'CurrencyKey', 'SalesTerritoryKey',
                    'SalesOrderID', 'SalesOrderLineNumber','OrderQty', 'UnitPrice', 'LineTotal', 'UnitPriceDiscount', 'DiscountAmount',
                    'ProductStandardCost', 'TotalProductCost', 'SubTotal', 'TaxAmt', 'Freight', 'PurchaseOrderNumber','CarrierTrackingNumber', 'OrderDate', 'DueDate', 'ShipDate']]

result = result.rename(columns={'SalesOrderID': 'SalesOrderNumber', 
                                'OrderQty': 'OrderQuantity', 
                                'UnitPriceDiscount': 'UnitPriceDiscountPct', 
                                'LineTotal': 'ExtendedAmount',
                                'SubTotal': 'SalesAmount',
                                'PurchaseOrderNumber':'CustomerPONumber'})

result
```

## LOAD

```python
db_op, db_etl = connect_databases()
result.to_sql('FactResellerSales', db_etl, if_exists='replace', index=False)
```
